```python
from pyspark.sql import SparkSession

# Create Spark Session
spark = SparkSession.builder \
    .appName("PostgresPySpark") \
    .config("spark.jars.packages", "org.postgresql:postgresql:42.7.4") \
    .enableHiveSupport() \
    .getOrCreate()

# PostgreSQL connection
jdbc_url = "jdbc:postgresql://host.docker.internal:5432/tpc"

properties = {
    "user": "postgres",
    "password": "tiger123",
    "driver": "org.postgresql.Driver"
}

# Create schema/database
spark.sql("CREATE DATABASE IF NOT EXISTS prod_db")

# Set schema
spark.sql("USE prod_db")

tables = ["customer","lineitem","nation","orders","part","partsupp","region","supplier"]

# Read PostgreSQL tables
for t in tables:

    df = spark.read.jdbc(
        url=jdbc_url,
        table=t,
        properties=properties
    )
    try:
        df.write.mode("overwrite").saveAsTable(f"prod_db.{t}")
        print(f"Loaded table: {t}")
    # Save into Spark catalog
    except:
        print(f"spark_catalog already created")
        continue
```

```python
#Aggregation
spark.sql(""" 
    select o_orderkey, sum(o_totalprice) over(partition by o_orderpriority order by o_totalprice) as ttl from orders 
""").show()
```

```python

spark.sql("""
with avg_cte as(
    select o_orderkey, Avg(o_totalprice) over(partition by o_orderpriority order by o_totalprice) as ttl from orders 
)
select o_orderkey, round(ttl,2) as avg_price from avg_cte
""").show()
```

```python
spark.sql(""" 
with rn_cte as(
    select c.c_name, o.o_orderkey, row_number() over(partition by o.o_orderpriority order by c.c_nationkey) as rn from orders o
    join customer c on c.c_custkey = o.o_custkey
)
select * from rn_cte
where rn <= 5
""").show()
```

```python
#table get the top 3 spending customers per day. The orders table schema is shown below:

spark.sql(""" 
with rn_cte as(
    select o_custkey, o_orderdate,row_number() over(partition by  o_orderdate order by  o_totalprice desc) as rn 
    from orders 
)
select * from rn_cte
where rn <=3
limit 5
""").show()

```

```python
#raking the fucntion

spark.sql(""" 
with rn_cte as(
    select o_custkey, o_orderdate,row_number() over(partition by  o_orderdate order by  o_totalprice desc) as rn, 
    rank() over(partition by  o_orderdate order by  o_totalprice desc) as rnk, 
    dense_rank() over(partition by  o_orderpriority order by  o_totalprice desc) as drnk
    from orders 
)
select * from rn_cte
where rn <=5
limit 20
""").show()
```

```python
# Write a query on the orders table that has the following output:
# o_custkey
# order_month: In YYYY-MM format, use strftime(o_orderdate, ‘%Y-%m’) AS order_month
# total_price: Sum of o_totalprice for that month
# three_mo_total_price_avg: The 3 month (previous, current & next) average of total_price for that customer
```

```python
# sum(o_totalprice), o_custkey, o_orderdate -> order_month, c_avg_three_month_price -> prev, current, next
```

```python
spark.sql(""" 

    select o.o_custkey,o.o_orderdate, avg(o.o_totalprice), sum(o.o_totalprice) over(partition by o.o_custkey) from orders o
    join customer c on o.o_custkey= c.c_custkey

""").show()
```
