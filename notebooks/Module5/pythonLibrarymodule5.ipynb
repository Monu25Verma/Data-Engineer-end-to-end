```python
#Read and Display Local File Contents. Read the previously saved JSON file and print its name and id.

import json
import requests

url = "https://pokeapi.co/api/v2/pokemon/1"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()

    #EXTRACT the requied field
    pokenmon_data = {
        "id" : data['id'],
        "name" : data['name']
    }
    print(pokenmon_data)
    #save to local in .json file
    with open("pokemon_api.json", "w") as f:
        json.dump(pokenmon_data, f, indent = 4)

    print("Successfully written in pokemon_api.json")
else:
    print("No data found with status code: {response.status_code}")
    
    
```

```python
#Parse Data and Insert into SQLite Database. Extract specific Pokemon attributes and store them in a SQLite database.

import requests
import psycopg2

# Fetch Pokemon data
response = requests.get("https://pokeapi.co/api/v2/pokemon/2")

if response.status_code == 200:

    data = response.json()

    pokemon_id = data["id"]
    pokemon_name = data["name"]
    pokemon_height = data["height"]
    pokemon_weight = data["weight"]

    try:
        conn = psycopg2.connect(
            host="host.docker.internal",
            database="pokemonApi",
            user="postgres",
            password="tiger123",
            port="5432"
        )

        cursor = conn.cursor()

        cursor.execute("""
        CREATE TABLE IF NOT EXISTS pokemon_api (
            id INTEGER PRIMARY KEY,
            name VARCHAR(50),
            height INTEGER,
            weight INTEGER
        )
        """)

        cursor.execute("""
        INSERT INTO pokemon_api (id, name, height, weight)
        VALUES (%s, %s, %s, %s)
        ON CONFLICT (id)
        DO UPDATE SET
            name = EXCLUDED.name,
            height = EXCLUDED.height,
            weight = EXCLUDED.weight
        """, (
            pokemon_id,
            pokemon_name,
            pokemon_height,
            pokemon_weight
        ))

        conn.commit()

        cursor.execute("SELECT * FROM pokemon_api")

        for row in cursor.fetchall():
            print(row)

        cursor.close()
        conn.close()

    except Exception as e:
        print("Database Error:", e)
```
