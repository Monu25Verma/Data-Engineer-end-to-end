```python

from pyspark.sql import SparkSession
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
schema = StructType([
    StructField("customer_id", IntegerType(), True),
    StructField("customer_name", StringType(), True),
    StructField("gender", StringType(), True),
    StructField("age", IntegerType(), True),
    StructField("product", StringType(), True),
    StructField("purschase_amount", StringType(), True)
])
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("test").getOrCreate()

data = spark.read.option("header", 'true') \
        .option("InferSchema", 'true') \
        .schema(schema) \
        .csv("module6data.csv")
data.show()
```

```python
## Question: How do you remove duplicate rows based on customer ID in PySpark?

data_unique = data.dropDuplicates()
data_unique.show()
```

```python
## Question: How do you handle missing values by replacing them with 0 in PySpark?
data_cleaning = data_unique.select(
    col("customer_id"),
    col("customer_name"),
    col("gender"),
    coalesce(col("age"), lit(0)).alias('age'),
    col("product"),
    when(col("purschase_amount") == "na", lit(0))
    .otherwise(col("purschase_amount"))
    .alias("purchase_amount")
)
data_cleaning.show()
```

```python
#
# Question: How do you remove outliers (e.g., age > 100 or purchase amount > 1000) in PySpark?
data_cleaning_outliers = data_cleaning.filter((col("age") <= 100) & (col("purschase_amount") <= 1000))
data_cleaning_outliers.show()
```

```python
#
# Question: How do you convert the Gender column to a binary format (0 for Female, 1 for Male) in PySpark?
data_cleaning_gender = data_cleaning_outliers.withColumn("Gender_binary", when(col("gender") == "Female", 0).otherwise(1))
data_cleaning_gender.show()
```

```python
## Question: How do you split the Customer_Name column into separate First_Name and Last_Name columns in PySpark?
data_cleaning_name = data_cleaning_gender.select(
    col("customer_id"),
    split(col("customer_name"), " ").getItem(0).alias("Fist_name"),
    split(col("customer_name"), " ").getItem(1).alias("Last_name"),
    col("age"),
    col("gender"),
    col("product"),
    col("purchase_amount")
)
           
data_cleaning_name.show()



```

```python
## Question: How do you calculate the total purchase amount by Gender in PySpark?
total_purchase_gender = data_cleaning_name.groupBy("gender") \
                .agg(sum("purchase_amount")) \
                .collect()

total_purchase_gender
```

```python
#Question: How do you print the results for total purchase amount by Gender and average purchase amount by Age group in PySpark?

average_purchase_by_age_group = data_cleaning_name.withColumn(
"Age_Group",
    when((col("age") >= 18) & (col("age") <= 30), "18-30")
    .when((col("age") >= 31) & (col("age") <= 40), "31-40")
    .when((col("age") >= 41) & (col("age") <= 50), "41-50")
    .when((col("age") >= 51) & (col("age") <= 60), "51-60")
    .otherwise("61-70")
).groupBy("Age_Group") \
    .agg(avg("purchase_amount").alias("average_purchase_amount")) \
    .collect()
average_purchase_by_age_group
```
